```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Lasso
from sklearn import metrics
```

```python
df= pd.read_csv("car data.csv")
```

__Encoding the categorical data__

```python
df.replace({'Fuel_Type':{'Petrol':0,'Diesel':1,'CNG':2}},inplace=True)

df.replace({'Seller_Type':{'Dealer':0,'Individual':1}},inplace=True)

df.replace({'Transmission':{'Manual':0,'Automatic':1}},inplace=True)

df = df.infer_objects(copy=False)
```

```python
df.head()
```

__splitting the data and target__

```python
X = df.drop(['Car_Name','Selling_Price'],axis = 1)
Y = df['Selling_Price']
```

```python
print(X)
```

```python
print(Y)
```

__Splitting training and test data__

```python
 X_train , Y_train , Y_test , X_test = train_test_split(X , Y , test_size = 0.1 , random_state = 2)
```

__Model Training__

__1.Linear Regression__

```python
#loading the linear regression model

lin_reg_model = LinearRegression()
```

```python
print(X_train.shape)
print(Y_train.shape)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
lin_reg_model.fit(X_train,Y_train)
```

__Model Evaluation__

```python
# prediction of trainig data

training_data_prediction = lin_reg_model.predict(X_train)
```

```python
#R squared error

error_score = metrics.r2_score(Y_train , training_data_prediction)
print("R squared error :",error_score)
```

__Visualization__

```python
plt.scatter(Y_train ,training_data_prediction )
plt.xlabel("Actual price ")
plt.ylabel("Predicted values")
plt.title("Actual prices vs predicted prices")
plt.show()
```

```python
# prediction of trainig data

test_data_prediction = lin_reg_model.predict(X_test)
```

```python
#R squared error

error_score = metrics.r2_score(Y_test , test_data_prediction)
print("R squared error :",error_score)
```

```python
plt.scatter(Y_test ,test_data_prediction )
plt.xlabel("Actual price ")
plt.ylabel("Predicted values")
plt.title("Actual prices vs predicted prices")
plt.show()
```
